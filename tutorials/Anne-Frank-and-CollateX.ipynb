### Compare text fragments

Next, we compare the two fragments from Anne Frank's manuscripts: 

```python
AF_collation = Collation()
AF_collation.add_plain_witness('Diary', 'Thinking now about my life in 1942, it all feels so unreal. That divine life was lived by a very different Anne Frank than the one who has now grown wise here. A divine life, that is what it was. Five admirers on every corner, 20 or more friends and acquaintances, the favorite of most of the teachers, pampered from top to bottom by Father and Mother, lots of candy, enough money, what more could you ask for?')
AF_collation.add_plain_witness('BH', 'Thinking about my life in 1942, it feels like an unreality. That life was lived by a very different Anne than the one being brought up in the Back House. Seen from here, it seems like being at home, on the Merry was so lovely, lots of friends and girlfriends, pampered by Father and Mother, lots of candy, enough money, what more could you ask for?')
result = collate(AF_collation, output='html', layout='vertical')
print(result)
```
